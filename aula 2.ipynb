3) Criar um algoritmo que leia um conjunto de informações (nome, sexo, idade, peso e
altura) dos atletas que participaram de uma olimpíada, e informar: a) o atleta do sexo
masculino mais alto; b) a atleta do sexo feminino mais pesada; c) a media de idade dos
atletas. O número de informações a ser lido é livre até que alguma condição de parada
seja definida

```python
## Entao, eu nao sei se tinha forma mais facil de fazer kkk, mas se esta funcioando eu estou no lucro kkkk..0

part = int(input('Digite numero de participante: '))
somaIdade = 0
atleta1 = ''
atleta2 = ''
maisAlto = 0
maisPesado = 0

for i in range(part):

  nome = input('Digite nome do participante: ')
  sexo = input('Digite sexo do participante(H|M): ')
  idade = int(input('Digite a idade do participante: '))
  peso = float(input('Digite o peso do participante: '))
  altura = float(input('Digite altura do participante: '))

  #a) atleta do sexo masculino mais alto
  if (sexo == 'H' or sexo == 'h') and (altura > maisAlto):
    atleta1 = nome
    maisAlto = altura
    #print('teste  --  ',atleta1)

  #b) atleta do sexo feminino mais pesada
  if (sexo == 'm' or sexo == 'M') and (peso > maisPesado):
    atleta2 = nome
    maisPesado = peso
    #print('teste  --  ',atleta2)
  somaIdade += idade

#c) média de idade dos atletas
mediaIdade = somaIdade / part

print(f'atleta do sexo masculino mais alto é {atleta1}')
print(f'atleta do sexo feminino mais pesada é {atleta2}')
print(f'média de idade dos atletas é {mediaIdade}')

```

4) Criar um algoritmo que leia um número (num) e imprima a soma dos números múltiplos
dé 5 no intervalo aberto entre 1 e num. Suponha que num será maior que zero

```python
num = int(input('Digite qualquer numero: '))
soma = 0

if num > 0:
  for i in range(num):
    if i % 5 == 0:
      soma += i
      print(f'soma: {soma}')
else:
  print('Digite um numero maior que 0')
```
