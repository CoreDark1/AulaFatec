A Suíça é o país com o maior IDH em 2021. Filtre a tabela *df* e atribua ao nome *sw* a cópia das informações desse país. Visualize suas 5 primeiras linhas.

**Dica**: Suíça em inglês é *Switzerland*.

```python
sw = df.loc[df['Entity'] == 'Switzerland']
sw.head()
```

Delete as colunas *Entity* e *Code* e atualize o nome *sw*.

```python
sw = sw.drop(columns=['Entity','Code'])
```

Reinicie os índices de *sw*. Elimine a coluna com os índices antigos , atualize *sw* e vejas as últimas 5 linhas.

```python
sw = sw.reset_index(drop=True)
sw.head()
```

Assim como fizemos no Brasil, iremos analisar quais foram os maiores fatores associados a óbitos na Suíça. Ordene a linha de 2019 em ordem decrescente.

```python
sw.iloc[29].sort_values(ascending=False)
```

Plote o gráfico de barras horizontal dos dados ordenados do item acima.

```python
sw.iloc[29].sort_values(ascending=False).plot(kind='barh',
        grid=True,
        )
```

Plote o gráfico mostrando a evolução no decorrer dos anos dos 5 maiores fatores associados à óbitos na Suíça em 2019.

```python
sw.plot(kind='line',
        x='Year',
        y=['high systolic blood pressure','smoking','high fasting plasma glucose','high ldl cholesterol','high body-mass index'],
        grid=True,
        figsize=(12,7)
        )
```

Os 5 maiores fatores associados à óbitos no Brasil foram:

Os 5 maiores fatores associados à óbitos na Suíça foram:
