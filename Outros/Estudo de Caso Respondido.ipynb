Para converter a Reais os preços, basta multiplicar o valor em rupia indiana pelo fator de conversão:

> $\text{valor em reais} = \text{valor em rupia indiana} \times \text{fator de conversão}$

Assim, se estivéssemos no dia 04 de setembro e quiséssemos converter 100 Rupias Indianas para reais, bastava fazer:

> $\text{valor em reais} = \text{valor em rupia indiana} \times \text{fator de conversão}$
>
> $\text{valor em reais} = 100 \times 0.05968$
>
> $\text{valor em reais} = 5.968 \text{ reais}$

Contudo, na tabela *fatcon* existem 30 entradas. Lembrando das aulas de Estatística Aplicada a Ciência de Dados, você resolveu calcular a média aritmética da coluna *fc* para encontrar um valor central que represente essa distribuição de valores:

```python
media = fatcon['fc'].mean()
media
```

Para verificar se esses valores não estão muito "espalhados" em torno da média aritmética, você calculou a variância dessa <ins>amostra</ins>:

```python
variancia = fatcon['fc'].var()
variancia
```

e também aproveitou para calcular seu desvio padrão:

```python
desvio = fatcon['fc'].std()
desvio
```

Após obter esses valores da média aritmética, variância e desvio padrão amostrais você ficou na dúvida: Será que eles estão indicando que esses fatores de conversão estão concentrados próximo à média aritmética ou será que estão muito dispersos dela?
```python
# Execute esta célula. Não altere as informações contidas nela. ################
def tobe():
    return np.random.choice(['é', 'não é'])

def tobe2():
    return np.random.choice(['seja', 'não seja'])

def andor():
    return np.random.choice(['e', 'ou'])

def hasnot():
    return np.random.choice(['tem', 'tão tem'])

def ineq1():
    l = ['maior que', 'menor que', 'maior ou igual a',
         'menor ou igual a', 'igual a']
    return np.random.choice(l)

def ineq2():
    l = ['mais que', 'menos que', 'mais que ou igual a',
         'menos que ou igual a', 'igual a']
    return np.random.choice(l)
################################################################################
if col1 == 'marca':
    brand = np.random.choice(sf2['marca'].unique())
    condm = tobe()
    connector = andor()
    text1 = f'A marca {condm} {brand} {connector} '

elif col1 == 'modelo':
    model = np.random.choice(sf2['modelo'].unique())
    condm = tobe()
    connector = andor()
    text1 = f'O modelo {condm} {model} {connector}'

elif col1 == '5g':
    condm = tobe().capitalize()
    connector = andor()
    text1 = f'{condm} 5g {connector} '

elif col1 == 'nfc':
    condm = hasnot().capitalize()
    connector = andor()
    text1 = f'{condm} NFC {connector} '

elif col1 == 'processador':
    condm = np.random.choice(sf2['processador'].unique())
    condition = tobe()
    connector = andor()
    text1 = f'O processador {condition} {condm} {connector} '

elif col1 == 'velocidade(GHz)':
    condition = ineq1()
    value = np.round(np.random.randint(1200, 3221) / 1000, 2)
    connector = andor()
    text1 = f'A velocidade é {condition} {value} GHz {connector} '

elif col1 == 'bateria(mAh)':
    condition = ineq1()
    value = np.random.choice(sf2['bateria(mAh)'].unique())
    connector = andor()
    text1 = f'A bateria tenha capacidade {condition} {value} mAh {connector} '

elif col1 == 'expansível até(GB)':
    value = 2 ** np.random.choice(np.arange(6, 12))
    connector = andor()
    text1 = f'Seja expansível até {value} GB {connector} '

else:
    condition = ineq2()
    value = np.random.choice(sf2['preço(R$)'].unique())
    connector = andor()
    text1 = f'O valor seja {condition} {value} reais {connector} '

if col2 == 'marca':
    brand = np.random.choice(sf2['marca'].unique())
    condm = tobe2()
    text2 = f'a marca {condm} {brand}.'

elif col2 == 'modelo':
    model = np.random.choice(sf2['modelo'].unique())
    condm = tobe2()
    text2 = f'o modelo {condm} {model}.'

elif col2 == '5g':
    condm = tobe2()
    text2 = f'{condm} 5g.'

elif col2 == 'nfc':
    condm = hasnot()
    text2 = f'{condm} NFC.'

elif col2 == 'processador':
    condm = np.random.choice(sf2['processador'].unique())
    condition = tobe()
    text2 = f'o processador {condition} {condm}.'

elif col2 == 'velocidade(GHz)':
    condition = ineq1()
    value = np.round(np.random.randint(1200, 3221) / 1000, 2)
    text2 = f'a velocidade seja {condition} {value} GHz. '

elif col2 == 'bateria(mAh)':
    condition = ineq1()
    value = np.random.choice(sf2['bateria(mAh)'].unique())
    text2 = f'a bateria tenha capacidade {condition} {value} mAh.'

elif col2 == 'expansível até(GB)':
    value = np.random.choice([64, 128, 256, 512, 1024, 2048])
    text2 = f'seja expansível até {value} GB.'

else:
    condition = ineq2()
    value = np.random.choice(sf2['preço(R$)'].unique())
    text2 = f'O valor seja {condition} {value} reais.'

print(text1 + text2)
print('''
Copie o texto acima e cole-o na célula abaixo.

(Atenção: Cada vez que você executar esta célula, valores distintos serão gerados.)
''')
```
