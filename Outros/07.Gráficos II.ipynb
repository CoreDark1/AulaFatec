# Gráficos II

Usando a biblioteca *pandas*, vimos como plotar e algumas formas de estilizar alguns tipos de gráficos. No encontro de hoje veremos como sobrepor gráficos e como salvar os gráficos plotados.

## Sobreposição de Gráficos

Existem situações em que desejamos inserir um gráfico no mesmo plano de outro gráfico. Isso é muito comum quando queremos comparar o comportamento de duas ou mais variáveis.

Usando o método *.plot* do módulo *pandas* podemos sobrepor todos os tipos de gráficos que vimos até o momento, <ins>**exceto gráfico de dispersão** ou **scatter plot**</ins>.

Usando o método *.plot*, para sobre por os gráficos, colocaremos o **nome das colunas contendo os valores que queremos comparar numa lista**.

Para entendermos melhor, usaremos a tabela a seguir como exemplo:

> [https://raw.githubusercontent.com/nightlySnoopy/database/main/temperaturas.csv](https://raw.githubusercontent.com/nightlySnoopy/database/main/temperaturas.csv)

Ela contém as temperaturas médias de 2 cidades coletadas no mesmo mês. Atribua essa tabela ao nome *dft* e calcule as estatísticas básicas dos valores presentes nela:

````python
import pandas as pd

patch = ("https://raw.githubusercontent.com/nightlySnoopy/database/main/temperaturas.csv")
dft = pd.read_csv(patch)
dft.describe()
````

## Sobreposição de Gráficos de Linhas

Suponha que queiramos comparar as temperaturas médias da *cidade A* com as da *cidade B*. O eixo horizontal receberá os valores dos dias. O que vai mudar agora será o argumento do parâmetro *y*: iremos colocar o nome das colunas numa <ins>lista</ins> que será o argumento desse parâmetro:

````python
dft.plot(kind='line',
x='dia',
y=['cidade A'])
````

Como temos dois gráficos, automaticamente é inserido a legenda com o nome das colunas para não haver confusão. De forma geral, o método *.plot( )* sempre tentará inserir a legenda no canto superior direito. Mas, caso isso atrapalhe a visualização, ele tentará inserir a legenda em outra posição que não atrapalhe.

Se usarmos o parâmetro

> label

podemos alterar os nomes que aparecem na legenda. Para isso, basta atribuir a ele uma lista com os nomes desejados <ins>**na mesma ordem do nome das colunas presentes na lista atribuída ao parâmetro y**</ins>.

````python
dft.plot(kind='line',
x='dia',
y=['cidade A', 'cidade B'],
label=['A', 'B'])
````

## Salvando os Gráficos Plotados

Após plotar o gráfico, podemos querer anexar a figura em algum tipo de documento (relatórios, trabalhos acadêmicos, entre outros.)

As duas principais extensões utilizadas para salvar arquivos de imagens são:

* PNG (Portable Nertwork Graphics): Apresenta alta capacidade de compressão, gerando imagens de alta qualidade e tamanho razoável aos padrões de internet atual. É muito utilizado quando se quer trabalhar com gráficos, logos e ilustrações. Além disso permite trabalhar com fundos transparentes.

* PDF (Portable Document Format): É um formato de arquivo que independe do dispositivo utilizado para sua leitura. De forma geral, se a imagem original for de alta qualidade, após ser convertida para PDF não ocorrerá distorções na imagem (O que não ocorre com o PNG). Quem utiliza LaTex, pode utilizar imagens no formato PDF na confecção de documentos.

Antes de salvar o arquivo com o gráfico, recomendo executar o seguinte procedimento:

1. Plotar o gráfico e verificar se está tudo certo: Título, nome dos eixos, entre outros;
2. Estando tudo certo, utilizaremos o método **.get_figure( )** seguido do método **.savefig( )**.

O método

<br/>

``` python
.get_figure( )
```
<br/>

não receberá argumento algum. Já o método

<br/>

``` python
.savefig( )
```
<br/>

receberá como argumento o string contendo o nome do arquivo com sua respectiva extensão.

Usaremos o exemplo das temperaturas médias das cidades *A* e *B* presentes na tabela **dft**:

````python
dft.plot(kind='line',
x='dia',
y=['cidade A', 'cidade B']).get_figure().savefig('temperatura.jpg')
````

Para visualizar o gráfico salvo no Colab:

1. Clique no ícone de pasta na barra lateral direita;
2. Clique com o botão direto em cima do nome do arquivo;
3. Clique em *Download*;
4. Selecione o nome do arquivo e o local onde ele será salvo.

## Qualidade Das Imagens

Um dos grande problemas de quando se trabalha com imagens é a sua qualidade.

Observe que a resolução da imagem do arquivo no formato PNG é baixa: parece meio "borrada". Caso você amplie a imagem, perceberá que isso irá piorar a situação. O que não acontece com os arquivos no formato PDF.

Isso ocorre porque o valor padrão do *pandas* para dpi (*Dots Per Inch* - Pontos por polegada) é 100. Para documentos, de forma geral, a resolução ideal é em torno de 150 a 300 dpi.

Podemos alterar o valor padrão acrescentando o parâmetro

> dpi

a atribuindo o valor desejado do dpi da figura.

Plote e salve a figura da gráfico de linha ta tabela *dfc*. Lembre-se que nessa tabela estão presentes as colunas

> Mês
>
> Etanol
>
> Gasolina

em seguida

1. salve em PDF
2. salve em PNG

Faça download das figuras e visualize-as.

````python
dfc.plot(kind='line',
x='Mês',
y=['Etanol', 'GasolinaC']).get_figure().savefig('preços1000.jpg', dpi=1000)
````

````python
dfc.plot(kind='line',
x='Mês',
y=['Etanol', 'GasolinaC']).get_figure().savefig('preços1000.pdf')
````

Altere a resolução para 300 dpi e use o formato png da mesma figura acima.

````python
dft.plot(kind='line',
x='dia',
y=['cidade A', 'cidade B']).get_figure().savefig('temperatura1.jpg', dpi=300)
````

# Arquivos no Formato *xlsx*

Desde que começamos a trabalhar com tabelas, usamos arquivos em **CSV**. Esse formato de arquivo, além de poder ser aberto em qualquer sistema operacional, tem outra vantagem:

> Importar arquivos no formato *CSV* é mais rápido e consome menos memória.

Por esse motivo, a maioria de dados arquivos que contêm dados são armazenados em *CSV*.

Mas, eventualmente encontramos arquivos no formato *XLSX*, ou seja, arquivos do Excel. Para ler esses tipos de arquivos, seguiremos os mesmos procedimentos utilizados para ler arquivos *CSV*. A única diferença é que iremos utilizar outra função:

> pd.read_csv( ) $\Longrightarrow$ pd.read_excel( )

Ou seja, basta trocar **csv** por **excel**.
