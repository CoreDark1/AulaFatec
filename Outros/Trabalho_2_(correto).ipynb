É muito comum em nossa sociedade associar a massa de um corpo com o tempo de sua queda.

Na tentativa de mostrar o equívoco deste pensamento, alguns cientistas fizeram o seguinte experimento:

1. Selecionaram um conjunto de esferas de aço inox maciço de diversos diâmetros e as pesaram;
2. Soltaram as esferas de uma altura de 40 m, medindo o tempo de queda usando sensores adequados para tal;
3. Repetiram esse mesmo experimento, mas dentro de uma câmara à vácuo;

Como as esferas foram selecionadas aleatoriamente, algumas delas apresentavam os mesmos diâmetros.

A tabela contendo essas informações se encontra no link abaixo:

> [https://raw.githubusercontent.com/nightlySnoopy/database/main/experiment.csv](https://raw.githubusercontent.com/nightlySnoopy/database/main/experiment.csv)
