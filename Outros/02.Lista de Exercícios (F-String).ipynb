2. Vimos que a operação de radiciação não é um comando "natural" do Python. Para resolvermos esa questão, podemos utilizar uma propriedade de potenciação/radiciação que nos diz o seguinte:

> $\displaystyle \sqrt[n]{x^{m}} = x^{\frac{m}{n}}$

Assim, para calcular a raiz quadrada de 9:

> $\displaystyle \sqrt{9} = \sqrt[2]{9^{1}} = 9^{\frac{1}{2}}$

ou seja, devemos executar $9^{\frac{1}{2}}$ no Python.

Para calcular a raiz quadrada de 10, você fará:

```python
# Substitua os ### pelo valor citado acima.

valor = 10

raiz = valor ** (1 / 2)
```

```python
# Digite os comandos nesta célula.

print(f'O valor da raiz quadrada de {valor} é {raiz}.')
```

3. Vimos em nosso primeiro encontro que o cálculo da montante em uma aplicação no regime de juros compostos é dada por:

> $\displaystyle \text{montante} = \text{capital} \times (1 + \text{taxa de juros})^{\text{tempo}}$

em símbolos:

> $\displaystyle M = C(1 + j)^{t}$

Suponha que você tenha R\$ 2 000.00 disponível para aplicar a uma taxa de juros de 0.819 \% ao mês durante 12 meses.

Na célula abaixo atribua os valores acima aos seus respectivos nomes e execute-a:

```python
# Substitua os ### pelos seus respectivos valores

capital = 2000

taxa = 0.819 # => Lembre-se que a taxa de juros deve estar no formato decimal.

tempo = 12

montante = capital * (1 + taxa) ** tempo

montante
```

Observe que o resultado da montante tem muitas casas decimais. Usando *f-string*, arredonde para 2 casas decimais o valor da *montante*.

```python
# Digite os comandos nesta célula.

text = f'A montante da aplicação foi de R$ {montante:.2f}.'

print(text)
```

4. Quando trabalhamos com números muito grandes ou muito pequenos, é comum utilizarmos a notação científica.

A 2ª Lei de Newton é conhecida pela função

> $\displaystyle F_{R} = ma$

Esta função nos informa a **intensidade** da força $F$, em newton (N), necessária para gerar uma aceleração $a$, em metro por segundo ao quadrado (m/s²), numa massa $m$, em quilograma (kg).

Suponha que um caminhão de 17 725 000 kg deva ser acelerado. Sabendo que a intensidade dessa aceleração deverá ser de 12 m/s², determine a força necessária para tal.

```python
# Substitua os ### pelos objetos adequados ####################################

massa = 17725000
aceleração = 12

força = massa * aceleração

imprimir = f'''
A força necessária para gerar uma aceleração de {aceleração} m/s² em um
caminhão com massa de {massa} kg é de {força} N.
'''

print(imprimir)
```

Observe que o valor da força obtido é muito grande. Na célula abaixo, usando *f-string*, faça o Python usar "_" como separador de milhar para a massa e para a força

```python
# Substitua os ### pelos comandos necessários.

text = f'''
A força necessária para gerar uma aceleração de {aceleração} m/s² em um
caminhão com massa de {massa:.2f} kg é de {força:.2f} N.
'''

print(imprimir)
```

Em algumas situações, valores muito grandes são melhor interpretados quando escritos em notação científica. Na celula abaixo, usando *f-string*, faça o Python imprimir a massa em notação científica com **4 casas decimais** e a força, com **2 casas decimais**.

```python
# Substitua os ### com os comandos necessários:

solution = f'''
A força necessária para gerar uma aceleração de {aceleração} m/s² em um
caminhão com massa de {massa:.4f} kg é de {força:.2f} N.'''

print(solution)
```

5. A lei de Coulomb diz que "A intensidade da força devido a interação entre duas cargas puntiformes é diretamente proporcional ao módulo do produto dessas cargas e inversamente proporcional ao quadrado de sua distância.":

> $\displaystyle F = \text{Constante Eletrostática} \times \frac{\left| \text{carga1} \times \text{carga2} \right|}{\text{distância}^{2}}$

ou, de forma mais simplificada:

> $\displaystyle F = k \frac{\left| q_{1} q_{2} \right|}{r^{2}}$

em que:
* $F$ é a intensidade da força em newton (N);
* $k$, no vácuo, vale $9.0 \times 10^{9}$ N $\cdot$ m²/C² (newton metro quadrado por coulomb quadrado);
* $q_{1}$ e $q_{2}$ são as cargas em coulomb (C);
* $r$ é a distância entre as cargas em metros (m).

Suponha que a carga $q_{1} = 9.168\times 10^{-9}$ C esteja a uma distância de 1.5 m da carga $q_{2} = -5.979 \times 10^{-6}$ C. Calcule a intensidade da força devido a interação dessas duas cargas na célula abaixo:

```python
# Substitua os ### pelos comandos adequados:

k = 9.0e9
q1 = 9.168e-9
q2 = 5.979e-6
r = 1.5

F = k * abs(q1 * q2) / r**2

print(f'A intensidade da força é {F} N.')
```

Faça o Pythom imprimir a mesma frase acima, mas em notação científica com 3 casas decimais na célula abaixo:
