7) Agora vamos adicionar todos os dataframes e gráficos que geramos no Passo 1, para
isso, vocês vão pegar o arquivo que fizeram download e enviar para o Colab, e chamar ele
dentro da célula do passo 6.


Confiram o nome que vocês deram para o arquivo, a célula completa com a primeira
aba preenchida fica da seguinte forma:

```python
%%writefile app.py
import streamlit as st
from aula5 import *
df_livros = pd.DataFrame(df.groupby('Categoria do Produto')['Preço'].agg(['count']))
grafico_vendas_livros = px.bar(
df_livros[['count']].sort_values('count', ascending=False),
x = 'count',
y = df_livros[['count']].sort_values('count', ascending=False).index,
text_auto = True,
title = 'Top Vendas por Categoria do Produto'
)
st.set_page_config(layout='wide')
st.title("Dashboard de Vendas :shopping_trolley:")
st.sidebar.title('Filtro de Vendedores')
filtro_vendedor = st.sidebar.multiselect(
'Vendedores',
df['Vendedor'].unique(),
)
if filtro_vendedor:
  df = df[df['Vendedor'].isin(filtro_vendedor)]
aba1, aba2, aba3, aba4 = st.tabs(['Dataset', 'Receita', 'Vendedores','Produtos'])
with aba1: st.dataframe(df)
with aba2:
  coluna1, coluna2 = st.columns(2)
  with coluna1:
    st.metric('Receita Total', format_number(df['Preço'].sum(), 'R$'))
    st.plotly_chart(grafico_map_estado, use_container_width=True)
    st.plotly_chart(grafico_rec_estado, use_container_width=True)
  with coluna2:
    st.metric('Quantidade de Vendas', format_number(df.shape[0]))
    st.plotly_chart(grafico_rec_mensal, use_container_width=True)
    st.plotly_chart(grafico_rec_categoria, use_container_width=True)
with aba3:
  coluna1, coluna2 = st.columns(2)
  with coluna1:
    st.plotly_chart(grafico_rec_vendedores, use_container_width=True)
  with coluna2:
    st.plotly_chart(grafico_vendas_vendedores, use_container_width=True)
with aba4:
  st.plotly_chart(grafico_vendas_livros, use_container_width=True)
```

Vejam que podemos ordenar os dados clicando pela coluna no dataframe gerado.
