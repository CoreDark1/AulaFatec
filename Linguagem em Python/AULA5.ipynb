7) Mas veja que temos um número todo quebrado e de difícil leitura, podemos então criar
uma função para deixar esse número mais fácil de ser lido. Sugestão: crie a função em
outra célula do Colab, assim caso precise alterar você executa somente a função e não
todo um longo trecho de código.

```python
def format_number(value, prefix = ''):
  for unit in ['', 'mil']:
    if value < 1000:
     return f'{prefix} {value:.2f} {unit}'
    value /= 1000
  return f'{prefix} {value:.2f} milhões'
```

8) Agora vamos chamar essa função para formatar o valor no passo 6.

```python
format_number(df['Preço'].sum(), 'R$')
```

10) O Pandas também permite que agrupemos as informações de forma fácil, e também é
importante ir salvando esses agrupamentos em novas variáveis para não perder o conjunto
original dos dados, que ainda vamos usar.
Estou agrupando então as informações pelo Local de compra e somando o preço para cada
local

```python
df_rec_estado = df.groupby('Local da compra')[['Preço']].sum()
#print(df_rec_estado)
```

11) Eu também posso escolher mostrar mais informações desse agrupamento, para isso
eu crio um subconjunto (subset) e passo as colunas que queiro separado por vírgula.
df_rec_estado = df.drop_duplicates(subset='Local da compra')[['Local da compra', 'lat', 'lo

```python
df_rec_estado = df.drop_duplicates(subset='Local da compra')[['Local da compra', 'lat', 'lon']]
#print(df_rec_estado)
```

12) Posso também fazer a função desse subconjunto com o local da compra, de forma a
mostrar o valor total da compra com as colunas que escolhi no passo 11, e também
ordenar os valores por alguma coluna, no caso escolhi o Preço.

```python
df_rec_estado = df.groupby('Local da compra')[['Preço']].sum()
df_rec_estado = df.drop_duplicates(subset='Local da compra')[['Local da compra', 'lat',
'lon']].merge(df_rec_estado, left_on='Local da compra', right_index=True).sort_values('Preço', ascending=False)
#print(df_rec_estado)
```

13) Vamos gerar um gráfico para facilitar essa visualização, vejam que eu separei as
colunas de latitude e longitude, então podemos colocar esses pontos no mapa, indicando
que as bolinhas maiores são referentes aos locais que tiverem maior quantidade de
vendas. Vocês podem dar zoom e interagir com o mapa.

```python
import plotly.express as px
grafico_map_estado = px.scatter_geo(
df_rec_estado, #estou passando o dataframe com meus dados
lat = 'lat', #coluna latitude
lon = 'lon', #coluna longitude
scope = 'south america', #estou restringindo meu mapa a américa do sul
size = 'Preço',
template = 'seaborn',
hover_name = 'Local da compra', #quando passar o nome em cima mostra o local
hover_data = {'lat': False, 'lon': False},
title = 'Receita por Estado'
)
grafico_map_estado
```

14) Vamos fazer agora um agrupamento quanto a receita mensal
freq = M (mês)
freq = A (ano)


```python
df_rec_mensal = df.set_index('Data da Compra').groupby(
    pd.Grouper(freq='M'))['Preço'].sum().reset_index()
#print(df_rec_mensal)
```

15) Podemos adicionar uma nova coluna no dataframe tambem fazendo
NOME_DF[‘nome da nova coluna’], assim podemos também adicionar o ano ao lado
dessa listagem

```python
df_rec_mensal['Ano'] = df_rec_mensal['Data da Compra'].dt.year
#print(df_rec_mensal)
```

16) E também podemos adicionar o mês, ter esses dados em uma coluna separada
facilita para gerar alguns gráficos


```python
df_rec_mensal['Mes'] = df_rec_mensal['Data da Compra'].dt.month_name()
#print(df_rec_mensal)
```

17) Agora podemos gerar um gráfico agrupando a receita mensal, o mês e o ano


```python
grafico_rec_mensal = px.line(
df_rec_mensal,
x = 'Mes',
y = 'Preço',
markers = True,
range_y = (0, df_rec_mensal.max()),
color = 'Ano',
line_dash = 'Ano',
title = 'Receita Mensal'
)
grafico_rec_mensal.update_layout(yaxis_title = 'Receita')
```

19) Também podemos gerar um gráfico de vendas por categoria, agrupado de forma

21) Até agora estávamos explorando os dados de vendas, mas temos informações de
vendedores também, podemos ver quais foram os que venderam mais, assim agrupamos
os vendedores por Preço.

```python
df_vendedores = pd.DataFrame(df.groupby('Vendedor')['Preço'].agg(['sum', 'count']).sort_values('sum',ascending=False))
print(df_vendedores)
```

22) E em seguida geramos o gráfico de 21

```python
grafico_rec_vendedores = px.bar(
df_vendedores[['sum']].sort_values('sum', ascending=False).head(7),
x = 'sum',
y = df_vendedores[['sum']].sort_values('sum', ascending=False).head(7).index,
text_auto = True,
title = 'Top 7 Vendedores por Receita'
)
grafico_rec_vendedores
```

23) E também podemos gerar a quantidade de vendas por vendedores utilizando o count
(contador), veja que antes estávamos somando por Preço, e vejam que não
necessariamente quem vendeu mais e quem teve a maior quantidade de vendas

```python
grafico_vendas_vendedores = px.bar(
df_vendedores[['count']].sort_values('count', ascending=False).head(7),
x = 'count',
y = df_vendedores[['count']].sort_values('count', ascending=False).head(7).index,
text_auto = True,
title = 'Top 7 Vendedores por Venda'
)
grafico_vendas_vendedores
```
