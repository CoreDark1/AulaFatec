6. Crie uma calculadora que receba a operação, o valor do numer 1, o valor do numero 2, e imprima o resultado. considere as 4 operaçoes basica.

```python
operacao = input('Digite sua operação (mais , menos , divisao, multiplicacao) ')
number1 = float(input('digite um numero '))
number2 = float(input('digite outro numero '))

if operacao == 'mais':
  resultado = number1 + number2
  print('o resultado da Adição é ', resultado)
elif operacao == "menos":
  resultado = number1 - number2
  print('o resultado da Subtração é ', resultado)
elif operacao == "divisao":
  resultado = number1 / number2
  print('o resultado da Divisão é ', resultado)
elif operacao == "multiplicacao":
  resultado = number1 * number2
  print('o resultado da Multiplicação é ', resultado)
else:
  print(' Modo invalido')
```

8. Faça um programa que receba o salário atual de um funcionário e calcule o valor do imposto que ele deve pagar conforme tabela abaixo:
 - Até 500 reais, não paga nada
 - De 500,01 até 1.000,00 - paga 5% de imposto
 - De 1000,01 até 1500,00 - paga 7% de imposto
 - Acima de 1500,01 - paga 10% de imposto

```python
salarioAtual = float(input('Digite seu Salario atual: '))

if salarioAtual <= 500:
  tot = salarioAtual
  print('Nao paga nada')
elif salarioAtual >= 500.01 and salarioAtual <= 1000.00:
  tot = salarioAtual * 0.05
  result = salarioAtual - tot
  print('Vai ser pago 5% do seu salario que corresponde à: ',tot)
  print('Seu salario atualizado ', result)
elif salarioAtual >= 1000.01 and salarioAtual <= 1500.00:
  tot = salarioAtual * 0.07
  result = salarioAtual - tot
  print('Vai ser pago 7% do seu salario que corresponde à: ',tot)
  print('Seu salario atualizado ', result)
else:
  tot = salarioAtual * 0.1
  result = salarioAtual - tot
  print('Vai ser pago 10% do seu salario que corresponde à: ',tot)
  print('Seu salario atualizado ', result)
```
