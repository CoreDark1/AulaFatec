1) Vamos implementar a cifra de Cesar vista em aula em Python.

Seu funcionamento é o seguinte:

· Passo 1: identificar o caractere dentro da sentença.

· Passo 2: achar a posição desse caractere dentro do alfabeto.

· Passo 3: pegar o caractere que está na posição + chave dentro do alfabeto.

· caso o (posição + chave) > 26, volte pra posição inicial.

· como o português possui caracteres com acento, vamos implementar um alfabeto um pouco maior, considerando esses casos.

· para decodificar uma mensagem, basta pegar o caractere na posição — chave dentro do alfabeto.

· Passo 4: criar uma nova sentença com os novos caracteres no lugar dos antigos.

· Passo 5: repetir até que o final da sentença seja atingido.

· Passo 6: retornar o resultado.

```python
# Solicitar o texto a ser encriptado ou decriptado
texto = 'XIWX'
# Chave a ser utilizada
chave = 4
# Determinar modo de operação (E = encriptar; D = decriptar)
modo = 'D'
# Conjunto de caracteres válidos no algoritmo
CARACTERES = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
# Variável para armazenar o texto criptografado (ou decifrado)
convertido = ''
# Converter todo o texto em maiúsculas
texto = texto.upper()
# Código que será executado em cada caractere do texto:
for caractere in texto:
  if caractere in CARACTERES:
    num = CARACTERES.find(caractere)
    if modo == 'E':
      num = num + chave
    if modo == 'D':
      num = num - chave
  # Manipular a rotação se o valor de num for maior do que o comprimento de CARACTERES ou menor que 0
    if num >= len(CARACTERES):
      num = num - len(CARACTERES)
    if num < 0:
      num = num + len(CARACTERES)
  # Adicionar (concatenar) o caractere correspondente a num na variável convertido
    convertido = convertido + CARACTERES[num]
  else:
# Concatenar o caractere sem efetuar criptografia ou decifragem
    convertido = convertido + caractere
# Mostrar o texto encriptado ou decifrado na tela
if modo == 'E':
  print('O texto criptografado é ', convertido)
elif modo == 'D':
  print('O texto decriptado é ', convertido)
else:
  print('Opção inválida')
```

2) Hash de Arquivos

· Crie um arquivo chamado a.txt com qualquer conteúdo textual dentro

· Crie outro arquivo chamado b.txt com o mesmo conteúdo textual dentro

· Faça upload desses arquivos no Google Colab e digite o código abaixo em uma célula

```python
import hashlib

def open_txt(file):
  with open(file) as f:
    return "".join(f.read())

file_1 = 'a.txt'
file_2 = 'b.txt'
text_1 = open_txt(file_1)
text_2 = open_txt(file_2)
def compare(text_1, text_2):

  hash_1 = hashlib.md5()
  hash_1.update(text_1.encode('utf-8'))
  hash_1 = hash_1.hexdigest()
  hash_2 = hashlib.md5()
  hash_2.update(text_2.encode('utf-8'))
  hash_2 = hash_2.hexdigest()

  return hash_1, hash_2;

hash = compare(text_1, text_2)
print(hash[0])
print(hash[1])
```

3) Criptografias Modernas

Essas criptografias que vimos em sala são ‘divertidas’ para troca de mensagens que não necessariamente precisam ser confidenciais, na vida real as empresas costumam utilizar o DES, AES ou RSA, por serem mais seguros e mais difíceis de serem quebrados usando força bruta.

A segurança do algoritmo RSA é baseado na dificuldade de fatorar números grandes em seus fatores primos. O processo de gerar chaves para o algoritmo RSA envolve a escolha de dois números primos grandes, e a multiplicação desses dois números para obter um número que será usado como módulo de criptografia.

Essa biblioteca já tem o RSA criado dentro dela, veja que ele gera uma chave privada e uma pública.

```python
from Crypto.PublicKey import RSA
chave= RSA.generate(2048) # gera uma chave com 2048 bits
chave_privada = chave.export_key()
chave_publica = chave.publickey().export_key()

from Crypto.Cipher import PKCS1_OAEP
menssagem = b'hello word'
chave_publica_obj = RSA.import_key(chave_publica)
cifra = PKCS1_OAEP.new(chave_publica_obj)
texto_encriptado = cifra.encrypt(menssagem)
texto_encriptado
```

Veja que a saída dele é bem diferente da cifra de César por exemplo, que o tamanho da saída era o tamanho do nosso texto.
